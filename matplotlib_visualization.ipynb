```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
y = 2*x
y
```

### Line Plot

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("X vs Y")
plt.show()
```

```python
plt.plot(x,y, color = 'g', linestyle = ':', linewidth = 5)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("X vs Y")
plt.show()
```

```python
y1 = 2*x
```

```python
y2 = 3*x
```

```python
plt.plot(x,y1, linestyle = ':')
plt.plot(x,y2, linewidth = 4)
plt.title("Two Lines in a Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1, color='r',linestyle = ':')

plt.subplot(2,1,2)
plt.plot(x,y2, linewidth = 4)

plt.title("Sub Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

### Bar Plot  (Categorical Value Distribution)

```python
student = {"Bob":87, "Matt":67, "Annie":100, "Julia":45}
```

```python
names = list(student.keys())
```

```python
names
```

```python
values = list(student.values())
```

```python
values
```

```python
plt.bar(names, values)
plt.title("Distributuin of Student Marks")
plt.xlabel("Names of Students")
plt.ylabel("Marks of Students")
plt.show()
```

```python
plt.barh(names, values)
plt.title("Distributuin of Student Marks")
plt.xlabel("Names of Students")
plt.ylabel("MArks of Students")
plt.grid(True)
plt.show()
```

### Scatter Plot

```python
x = [10,20,30,40,50,60,70,80,90]
y = [8,1,7,2,5,9,3,5,7]

plt.scatter(x,y)
plt.show()
```

```python
x = [10,20,30,40,50,60,70,80,90]
y = [8,1,7,2,5,9,3,5,7]

plt.scatter(x,y, marker="*", c='r', s = 100)
plt.show()
```

```python
x = [10,20,30,40,50,60,70,80,90]
y = [8,1,7,2,5,9,3,5,7]
z = [1,2,3,4,6,5,7,8,9]

plt.scatter(x,y, marker="*", c='r', s = 100)
plt.scatter(x,z, marker=".", c='y', s = 100)
plt.show()
```

```python
x = [10,20,30,40,50,60,70,80,90]
y = [8,1,7,2,5,9,3,5,7]
z = [1,2,3,4,6,5,7,8,9]

plt.subplot(1,2,1)
plt.scatter(x,y, marker="*", c='r', s = 100)

plt.subplot(1,2,2)
plt.scatter(x,z, marker=".", c='y', s = 100)

plt.show()
```

### Pie Chart (Distribution btw diff categorical values)

```python
fruit = ['Apple', 'Orange', 'Mango', 'Guava']
quantity = [67, 34, 100, 29]
```

```python
plt.pie(quantity, labels=fruit)
plt.show()
```

```python
plt.pie(quantity, labels=fruit, autopct='%0.1f%%', colors=['yellow', 'blue', 'green', 'red'])
plt.show()
```

### Doughnut Chart

```python
plt.pie(quantity, labels=fruit, radius=2)
plt.pie([100], radius=1, colors=['w'])
plt.show()
```
