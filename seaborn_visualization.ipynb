```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
fmri.shape
```

### Line Plot

```python
sns.lineplot(x='timepoint', y='signal', data=fmri)
plt.show()
```

```python
sns.lineplot(x='timepoint', y='signal', data=fmri, hue='event')
plt.show()
```

```python
sns.lineplot(x='timepoint', y='signal', data=fmri, hue='event', style='event')
plt.show()
```

```python
sns.lineplot(x='timepoint', y='signal', data=fmri, hue='event', style='event', markers=True)
plt.show()
```

### Bar Plot

```python
pokemon.head()
```

### Scatter Plot

```python
iris = pd.read_csv("Iris.csv")
iris.head()
```

```python
sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm', data=iris)
plt.show()
```

```python
sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm', data=iris, hue='Species', style='Species')
plt.show()
```

```python
sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm', data=iris, hue='PetalLengthCm')
plt.show()
```

### Histogram / Distplot

```python
diamond = pd.read_csv('diamonds.csv')
diamond.head()
```

```python
sns.distplot(diamond['price'])
plt.show()
```

```python
sns.distplot(diamond['price'], hist=False)
plt.show()
```

```python
sns.distplot(diamond['price'], color='g')
plt.show()
```

```python
sns.distplot(diamond['price'], color='g', bins=10, kde=False)
plt.grid(False)
plt.show()
```

```python
sns.distplot(diamond['price'], color='g', vertical=True)
plt.show()
```

### Joint Plot (Scatter Plot + Histogram)

```python
sns.jointplot(x='SepalLengthCm', y='PetalLengthCm', data=iris)
plt.show()
```

```python
sns.jointplot(x='SepalLengthCm', y='PetalLengthCm', data=iris, color='olive')
plt.show()
```

```python
sns.jointplot(x='SepalLengthCm', y='PetalLengthCm', data=iris, color='brown', kind='reg')
plt.show()
```

### Box Plot (Categorical(x) , numrical(y)) - MULTIVARIATE VIZLN

```python
churn = pd.read_csv('Telco-Customer-Churn.csv')
churn.head()
```

```python
sns.boxplot(x='Churn', y='tenure', data=churn)
plt.show()
```

```python
sns.boxplot(x='InternetService', y='MonthlyCharges', data=churn)
plt.show()
```

```python
sns.boxplot(x='Contract', y='tenure', data=churn, palette='Set1')
plt.show()
```

```python
sns.boxplot(x='Contract', y='tenure', data=churn, color='orange')
plt.show()
```

```python
sns.boxplot(x='Contract', y='tenure', data=churn, linewidth=3)
plt.show()
```

```python
sns.boxplot(x='Contract', y='tenure', data=churn, order=['Two year', 'Month-to-month', 'One year'])
plt.show()
```

```python
sns.boxplot(x='Contract', y='tenure', data=churn, hue='PaymentMethod', order=['Two year', 'Month-to-month', 'One year'])
plt.show()
```
